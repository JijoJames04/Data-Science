```python
#Python Imaging Library (PIL) and to display images in Jupyter notebook
from PIL import Image
from IPython.display import display

#And let's just look at the image 
im = Image.open('Screenshot_20221116_000058.png')
display(im)
```

```python
array = np.array(im)
print(array.shape)
array
```

```python
mask = np.full(array.shape, 255)
mask
```

```python
from numpy.core.memmap import uint8
modified_array = array - mask

#And let's convert all of the negative values to positive values
modified_array = modified_array * -1

#And as a last step, let's tell numpy to set the value of the datatype correctly
modified_array = modified_array.astype(uint8)
modified_array
```

```python
#And lastly, lets display this new array. We do this by using the fromarray() function in the python 
# imaging library to convert the numpy array into an object jupyter can render
display(Image.fromarray(modified_array))
```
